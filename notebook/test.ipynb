### 将时间戳转换为datetime格式

```python
df = pd.read_csv(data_path,
                 parse_dates=["Unnamed: 0"])
df.info()
```

```python
df.columns
```

```python
df  = df[['ActivePower', 'WindDirection','WindSpeed']]
df.dropna(axis=0, inplace=True)
df['WindDirection_sin'] = np.sin(df['WindDirection'])
df['WindDirection_cos'] = np.cos(df['WindDirection'])
df.describe()
```

```python
# duplicate the date column to change it's name 
df['DateTime'] = df['Unnamed: 0'] 
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df['DateTime'].head(20)
```

```python
# Add datetime parameters 
df['DateTime'] = pd.to_datetime(df['DateTime'], 
 format = '%Y-%m-%dT%H:%M:%SZ', 
 errors = 'coerce')

df['year'] = df['DateTime'].dt.year
df['month'] = df['DateTime'].dt.month
df['day'] = df['DateTime'].dt.day
df['hour'] = df['DateTime'].dt.hour
df['minute'] = df['DateTime'].dt.minute
```

```python
# Drop original DateTime column
df.drop('DateTime', axis=1, inplace= True)
```

```python
df.columns
```

```python
df.isna().sum()
```

### 填充缺失值

```python
from tqdm import tqdm

# Fill numeric rows with the median
for label, content in tqdm(df.items()):
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            # Add a binary column which tells if the data was missing our not
            df[label+"_is_missing"] = pd.isnull(content)
            # Fill missing numeric values with median since it's more robust than the mean
            df[label] = content.fillna(content.median())
```

```python
df.isna().sum()
```

```python
# Turn categorical variables into numbers
for label, content in tqdm(df.items()):
    # Check columns which aren't numeric
    if not pd.api.types.is_numeric_dtype(content):
        # Add binary column to inidicate whether sample had missing value
        df[label+"_is_missing"] = pd.isnull(content)
        # We add the +1 because pandas encodes missing categories as -1
        df[label] = pd.Categorical(content).codes+1
```

## 随机森林筛选特征

```python
from sklearn.ensemble import RandomForestRegressor

X = df.drop("ActivePower", axis=1)
y = df['ActivePower']

# 创建随机森林分类器
rf_classifier = RandomForestRegressor(n_estimators=100, max_depth=6, random_state=42)

# 拟合模型
rf_classifier.fit(X, y)

# 获取特征重要性
feature_importances = rf_classifier.feature_importances_
feature_importances
```

```python

# 定义特征重要性阈值
threshold = 0.1  # 根据需要自行调整

# 根据特征重要性选择特征
selected_features = np.where(feature_importances > threshold)[0]

# 打印选定的特征索引
print("选定的特征索引:", selected_features)

# 根据选定的特征索引获取选定的特征名字
selected_feature_names = [df.columns[i] for i in selected_features]

# 打印选定的特征名字
print("选定的特征名字:", selected_feature_names)
```
